```python
with open("wizard_of_oz.txt", "r", encoding="utf-8") as f:
    text = f.read()
chars = sorted(set(text))
# set tạo 1 danh sách các ký tự duy nhất trong văn bản, sau đó sorted sắp xếp chúng theo thứ tự bảng chữ cái   
print(chars)
vocb_size = len(chars)
```

```python
str_to_int={ch:i for i,ch in enumerate(chars)}
int_to_str={i:ch for i,ch in enumerate(chars)}
# Tạo hai dictionary str_to_int và int_to_str để ánh xạ giữa ký tự và số nguyên. 
# str_to_int ánh xạ mỗi ký tự đến một số nguyên duy nhất, 
# trong khi int_to_str ánh xạ ngược lại từ số nguyên đến ký tự. 
# Cả hai dictionary này được tạo bằng cách sử dụng enumerate để gán một số nguyên duy nhất cho mỗi ký tự trong danh sách chars.
print(str_to_int)
print(int_to_str)
```

ví dụ chars = ['e', 'h', 'l', 'o'] 
str_to_int sẽ là: {'e': 0, 'h': 1, 'l': 2, 'o': 3}
    Sử dụng: str_to_int['h'] trả về 1.
int_to_str sẽ là: {0: 'e', 1: 'h', 2: 'l', 3: 'o'}
    Sử dụng: int_to_str[1] trả về 'h'.

```python
encode = lambda s: [str_to_int[c] for c in s]
decode = lambda l: "".join([int_to_str[i] for i in l])
# encode là một hàm lambda để mã hóa một chuỗi s thành một danh sách các số nguyên bằng cách sử dụng dictionary str_to_int.
# decode là một hàm lambda để giải mã một danh sách các số nguyên l thành một chuỗi bằng cách sử dụng dictionary int_to_str.
print(encode("hello world"))
print(decode(encode("hello world")))
```

```python
data = torch.tensor(encode(text), dtype=torch.long)
print(data.shape, data.dtype)
print(data[:100])
```

```python
indices = data[:100].tolist() # Chuyển đổi tensor data thành một danh sách các số nguyên bằng cách sử dụng phương thức tolist().
print(indices)
original = decode(indices) # Giải mã danh sách các số nguyên indices thành một chuỗi bằng cách sử dụng hàm decode.
print(original)
```

```python
n = int(0.8 * len(data))    # Chia dữ liệu thành hai phần: 80% cho tập huấn luyện (train_data) và 20% cho tập kiểm tra (val_data).
train_data = data[:n]       
val_data = data[n:]
```

```python
block_size = 8

x=train_data[:block_size]
y=train_data[1:block_size+1]

for t in range(block_size):
    context = x[:t+1]
    target = y[t]   
    print(f"when input is {context} the target: {target}")
    
```
