```python
import torch
import time
import numpy as np

size = 10000
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Khởi tạo
a_torch = torch.rand(size, size).to(device)
b_torch = torch.rand(size, size).to(device)
a_np = np.random.rand(size, size).astype(np.float32)
b_np = np.random.rand(size, size).astype(np.float32)

# 1. Đo PyTorch (Matrix Multiplication)
if device == 'cuda':
    torch.cuda.synchronize() # Đợi GPU sẵn sàng
start = time.time()

result_torch = torch.matmul(a_torch, b_torch)

if device == 'cuda':
    torch.cuda.synchronize() # Đợi GPU làm xong mới bấm dừng đồng hồ
print(f"PyTorch ({device}): {time.time() - start:.4f}s")

# 2. Đo NumPy (Matrix Multiplication)
start = time.time()
result_np = np.matmul(a_np, b_np)
print(f"NumPy (CPU): {time.time() - start:.4f}s")
```
