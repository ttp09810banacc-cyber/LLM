bias là false thì là weight*x
if bias==False and input ==[0,0,0] --> output=[0,0,0]

```python
import torch.nn.functional as F

tensor1 = torch.tensor([1.0,2.0,3.0])   #tổng là 1

softmax_output = F.softmax(tensor1, dim=0)

softmax_output
```

```python
tensor2 = torch.tensor([[1.0, 2.0, 3.0], 
                        [1.0, 1.0, 1.0]])
softmax_output_1 = F.softmax(tensor2, dim=1) #dim=1 thì tổng ngang = 1 dim=0 thì tổng dọc là 1

softmax_output_1
```
